```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1000,1000,num=400) #score diff
y = 1/(1+10**(x/500))
plt.plot(x,y)
plt.title("Win Probability Function")
plt.xlabel("R$_{b}$ - R$_{a}$")
plt.ylabel("Win Probability of Player A")
plt.show()
```

```python
x = np.linspace(2000,2700,num=700) #score diff
y = 0.8 + (x-1000)/10000
plt.plot(x,y,[1700,2000],[0.90,0.90],[2700,3000],[0.97,0.97], color = '#FF4500')
plt.title("Smoothing parameter ")
plt.xlabel("g$_{t}$")
plt.ylabel("$_{t}$")
plt.show()
```

```python
# Example data: probability forecasts and corresponding Brier Scores
probability_forecasts = np.linspace(0,1,100)
log_scores = -np.log(probability_forecasts)

plt.figure(figsize=(10, 6))
plt.plot(probability_forecasts, log_scores, color='b')
plt.title('Ignorance Score vs Probability Forecast')
plt.xlabel('Win Probability')
plt.ylabel('Brier Score')
plt.xlim(0, 1)  # Set x-axis limits to the range of probabilities
plt.ylim(0, 4)  # Set y-axis limits to the range of Brier Scores
plt.grid(True)
plt.show()
```

```python
# Example data: probability forecasts and corresponding Brier Scores
probability_forecasts = np.linspace(0,1,100)
brier_scores = (1 - probability_forecasts)**2

plt.figure(figsize=(10, 6))
plt.plot(probability_forecasts, brier_scores, color='b')
plt.title('Brier Score vs Probability Forecast')
plt.xlabel('Win Probability')
plt.ylabel('Brier Score')
plt.xlim(0, 1)  # Set x-axis limits to the range of probabilities
plt.ylim(0, 1)  # Set y-axis limits to the range of Brier Scores
plt.grid(True)
plt.show()
```
